Metadata

Date - Trade Data: Represents the date of the trading data, indicating when the stock market activity occurred.

**Symbol** - Name of Stock: Refers to the unique code or symbol assigned to a particular stock. It is essentially the shorthand representation of the company's name on the stock exchange.

**Series** - Type of Security: Specifies the type of security, which could include different classes of financial instruments such as equity shares, preference shares, or other financial products.**(EQ= Equity)**

**Prev Close** - Previous Day's Closing Price: Indicates the closing price of the stock on the previous trading day.

**Open** - Opening Price for the Day: Represents the initial price at which a stock is traded on a given day.

**High**- Highest Price for the Day: Denotes the highest trading price reached by the stock during the trading day.

**Low** - Lowest Price for the Day: Represents the lowest trading price reached by the stock during the trading day.

**Last** - Last Trade Price: Refers to the price at which the last trade was executed.

**Close** - Closing Price: Indicates the final trading price of the stock at the end of the trading day.

**VWAP** - Volume-Weighted Average Price: VWAP is a ratio of the cumulative share price to the cumulative volume traded over a given time period. It provides insight into the average price at which a stock is traded, weighted by the volume of trades.

**Volume** - Volume Traded for the Day: Represents the total number of shares or contracts traded during a specific time period, typically a trading day.

**Turnover** - Turnover Ratio: The turnover ratio is the ratio of sellers to buyers of a stock. It helps in understanding the market activity and liquidity.

**Trades** - Number of Trades: Indicates the total number of trades (buy and sell transactions) that occurred during the trading day.

**Deliverable Volume **- Amount of Deliverable Volume: Represents the volume of shares that were actually delivered (transferred) as opposed to being traded intraday.

**% Deliverable**  - Percentage of Deliverable Shares: Indicates the percentage of shares that were delivered out of the total traded volume.


# MOST PREFFERED
For time series classification, we can consider the following algorithms:

Distance-based Approaches: These include k-nearest neighbors (KNN) and support vector machines (SVM). They measure the distance between the test object and all of the objects in the training data set1.

Dynamic Time Warping (DTW): This is a distance-based algorithm used for measuring the distance between two time series1.

Deep Learning Models: These include Recurrent Neural Networks (RNN), Long Short-Term Memory (LSTM), and Convolutional Neural Networks (CNN). They are particularly useful for time series data due to their ability to capture temporal dependencies2.

# LESS PREFERRED
Hidden Markov Models (HMM): These are statistical models that assume there is an underlying process that is not observable (hidden) but can be inferred through another set of observed variables.

AutoRegressive Integrated Moving Average (ARIMA): This is a forecasting technique that projects the future values of a series based entirely on its own inertia.

Prophet: This is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects.

Remember, the choice of algorithm depends on your specific use case and the nature of your data. It’s always a good idea to try multiple algorithms and choose the one that performs best on your validation set.

# PROBABILITY



As for your requirement of getting class probabilities, most of these algorithms provide a method to get the class probabilities. For example, in scikit-learn, methods like **predict_proba** can be used to get the class probabilities.

# normalization or standardization

Please note that preprocessing steps like normalization or standardization might be necessary before feeding the data into these algorithms. Also, for time series data, feature engineering like creating lag features, rolling mean, etc., can be very useful.

# Ensemble

Lastly, for the ensemble part, you can use techniques like voting, stacking, or boosting. In voting, you take the majority vote of classifications from different models. In stacking, you use the predictions of several base models to build a new model. Boosting is an iterative technique that adjusts the weight of an observation based on the last classification. If an observation was classified incorrectly, it tries to increase the weight of this observation and vice versa. Boosting in general builds strong predictive models.

# In the context of time series data, there are three main classes of features that can be created1:

**Date Time Features:** These are components of the time step itself for each observation. For example, the year, month, day, and time could all be separate features.

**Lag Features: **
These are values at prior time steps. They can help the model capture temporal dependencies.

**Window Features:** These are summaries of values over a fixed window of prior time steps. For example, you could create features that capture the mean, minimum, maximum, or standard deviation of the values over the past week.

# what are rolling parameters (window size is being kept 5 because there is holiday on saturday and sunday)

For each data point, we'll take the average of VWAP values over the previous 2 days (including the current day). If there are not enough previous days to fulfill the window size, NaN (not a number) will be returned for that data point.

So, the calculation would be:

For 2024-01-01: There's only one data point, so the rolling mean would be NaN.
For 2024-01-02: (50 + 55) / 2 = 52.5
For 2024-01-03: (55 + 60) / 2 = 57.5
For 2024-01-04: (60 + 65) / 2 = 62.5
For 2024-01-05: (65 + 70) / 2 = 67.5

# what is lag

 lag_1, lag_2, and lag_3 refer to lag features created based on the VWAP (Volume Weighted Average Price) column. Let's break down what these lag features represent:

lag_1: This represents the VWAP value at the previous time step (one time step ago).
lag_2: This represents the VWAP value at two time steps ago.
lag_3: This represents the VWAP value at three time steps ago.
These lag features are created by shifting the VWAP column backward by one, two, and three time steps, respectively, using the shift() function. By doing this, you create features that capture the VWAP values from previous time steps. These lag features can be useful for capturing temporal dependencies and patterns in the data, which can then be utilized by machine learning models for making predictions or performing analysis.

```python

# Date Time Features
all_data['Year'] = all_data['Date'].dt.year
all_data['Month'] = all_data['Date'].dt.month
all_data['Day'] = all_data['Date'].dt.day
#all_data['Hour'] = all_data['Date'].dt.hour

# Lag Features
num_lags = 3  # for example, use the past 3 time steps
for lag in range(1, num_lags + 1):
    all_data[f'lag_{lag}'] = all_data['VWAP'].shift(lag)

# Window Features FOR vwap
window_size = 5  # for example, use a window size of 5
all_data['rolling_mean'] = all_data['VWAP'].rolling(window=window_size).mean()
all_data['rolling_min'] = all_data['VWAP'].rolling(window=window_size).min()
all_data['rolling_max'] = all_data['VWAP'].rolling(window=window_size).max()
all_data['rolling_std'] = all_data['VWAP'].rolling(window=window_size).std()



# Window Features
window_size = 5  # for example, use a window size of 2
all_data['rolling_mean1'] = all_data['Deliverable Volume'].rolling(window=window_size).mean()
all_data['rolling_min1'] = all_data['Deliverable Volume'].rolling(window=window_size).min()
all_data['rolling_max1'] = all_data['Deliverable Volume'].rolling(window=window_size).max()
all_data['rolling_std1'] = all_data['Deliverable Volume'].rolling(window=window_size).std()

# Drop rows with NaN values caused by lag and window features
all_data.dropna(inplace=True)
```

```python


# Construct GM for each model and calculate weights
similarity = np.zeros((len(base_models), len(test_pca)))
for m, model in enumerate(base_models):
    # Make predictions on the testing data
    model_predictions = predictions[m]

    # Construct GM, GM is a graph in which there is an edge connecting two test examples with an edge if and only if two examples are in the same cluster.
    GM = np.zeros((len(test_pca), len(test_pca)))
    for i in range(len(model_predictions)):
        for j in range(i + 1, len(model_predictions)):
            if model_predictions[i] == model_predictions[j]:  # Compare individual elements
                GM[i, j] = 1
                GM[j, i] = 1

    # Calculate weights
    for i in range(len(test_pca)):
        VM = np.where(GM[i] == 1)[0]
        VT = np.where(GT[i] == 1)[0]
        similarity[m, i] = len(np.intersect1d(VM, VT)) / (len(VM) + len(VT))
# Normalize weights
weights = similarity / np.sum(similarity, axis=0)
# Assuming weights is 2D numpy array of shape (3, 1423)
weight1 = weights[0:1, :]  # Extracts the first row
weight2 = weights[1:2, :]
weight3 = weights[2:3, :]  # Extracts the second row
probabilities1=y_prob1
probabilities2=y_prob3
probabilities3=y_prob4
import numpy as np


# weight1 has shape (1, 1423)
# probabilities1 has shape (1423, 2)

# Reshape weight1 to match the shape of probabilities1
weight1_reshaped = weight1.reshape((1, -1))

# Perform element-wise multiplication
new_probabilities1 = probabilities1 * weight1_reshaped.T

# Printing the shape and the new array
print("Shape of new_probabilities1:", new_probabilities1.shape)
print("New probabilities1:", new_probabilities1)


# Reshape weight1 to match the shape of probabilities1
weight2_reshaped = weight2.reshape((1, -1))

# Perform element-wise multiplication
new_probabilities2 = probabilities2 * weight2_reshaped.T


weight3_reshaped = weight3.reshape((1, -1))

# Perform element-wise multiplication
new_probabilities3 = probabilities3 * weight3_reshaped.T


# Printing the shape and the new array
print("Shape of new_probabilities2:", new_probabilities2.shape)
print("New probabilities2:", new_probabilities2)
import numpy as np

# Create the final_predictions matrix
final_probabilities = np.maximum.reduce([new_probabilities1, new_probabilities2, new_probabilities3])

# Printing the shape and the final predictions matrix
print("Shape of final_predictions:", final_probabilities.shape)
print("Final prrobabilities:", final_probabilities)

# Create final_predictions1 matrix
final_predictions1 = (final_probabilities[:, 0] < final_probabilities[:, 1]).astype(int)
#for example, if (final_predictions[:, 0] < final_predictions[:, 1]) then we get TRUE, which numerically means 1, which represents class T.

# Reshape final_predictions1 to match the desired shape
final_predictions1 = final_predictions1.reshape((1, -1))

# Printing the shape and the final_predictions1 matrix
print("Shape of final_predictions1:", final_predictions1.shape)
print("Final predictions1:", final_predictions1)
# Reshape final_predictions1 to match the shape of y_test
final_predictions1 = final_predictions1.reshape(-1)
print(final_predictions1)
print(final_predictions1.shape)

```
